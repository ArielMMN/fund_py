```python
# Criação da matriz 2x2
M = [[0, 0], [0, 0]]

# Leitura dos elementos da matriz
for i in range(2):
    for j in range(2):
        M[i][j] = int(input(f"Digite o elemento da linha {i} e coluna {j}: "))

# Encontrar o maior elemento da matriz
maior = M[0][0]
for i in range(2):
    for j in range(2):
        if M[i][j] > maior:
            maior = M[i][j]

# Construir a matriz resultante multiplicando cada elemento por 'maior'
matriz_resultante = [[0, 0], [0, 0]]
for i in range(2):
    for j in range(2):
        matriz_resultante[i][j] = M[i][j] * maior

# Exibir a matriz resultante
print("Matriz resultante:")
for i in range(2):
    for j in range(2):
        print("%3d" % matriz_resultante[i][j], end=" ")
    print()
```

```python
# Leitura do tamanho da matriz
N = int(input("Digite a ordem da matriz: "))

# Criação da matriz M com N linhas e N colunas
M = [[0] * N for _ in range(N)]

# Preenchendo a matriz
for i in range(N):
    for j in range(N):
        # O primeiro elemento é 1 (linha 0, coluna 0)
        if j == 0:
            M[i][j] = 1 << i  # 2 elevado a i
        else:
            M[i][j] = M[i][j - 1] * 2  # Dobra o valor anterior

# Impressão da matriz
print()
for i in range(N):
    for j in range(N):
        print("%4d" % M[i][j], end="")
    print()  # Nova linha
```

```python
import random

# Criação da matriz 12x12 preenchida com números aleatórios de 0 a 10
M = [[random.randint(0, 10) for _ in range(12)] for _ in range(12)]

# Leitura da operação a ser realizada (S ou M)
operacao = input("Digite 'S' para soma ou 'M' para média: ").strip().upper()

# Variável para armazenar a soma dos elementos acima da diagonal secundária
soma = 0
# Variável para contar a quantidade de elementos nessa área
contagem = 0

# Percorrendo a matriz para identificar elementos acima da diagonal secundária
for i in range(12):
    for j in range(12):
        if i + j < 11:  # Condição para área verde (acima da diagonal secundária)
            soma += M[i][j]
            contagem += 1

# Impressão da matriz formatada
print("\nMatriz:")
for linha in M:
    for elemento in linha:
        print("%3d" % elemento, end="")
    print()  # Nova linha

# Cálculo e exibição do resultado conforme operação
if operacao == 'S':
    print("\nSoma dos elementos acima da diagonal secundária:", soma)
elif operacao == 'M':
    media = soma // contagem  # Divisão inteira para média inteira
    print("\nMédia dos elementos acima da diagonal secundária:", media)
else:
    print("\nOperação inválida. Use apenas 'S' ou 'M'.")
```
